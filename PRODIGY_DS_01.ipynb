```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('train.csv')
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
# Impute missing 'Embarked' values with the mode (though not directly used for these plots, good practice)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# Drop 'Cabin' column (not directly used for these plots, good practice)
df = df.drop('Cabin', axis=1)

print("--- Visualizing Distributions ---")
```

```python
# 1. Histogram for the distribution of Age (continuous variable)
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], kde=True, bins=30)
plt.title('Distribution of Age in Titanic Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# 2. Bar chart for the distribution of Gender (categorical variable)
plt.figure(figsize=(6, 5))
sns.countplot(x='Sex', data=df, palette='viridis')
plt.title('Distribution of Gender in Titanic Passengers')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```
